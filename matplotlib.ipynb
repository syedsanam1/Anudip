```python
# np.exp() syntax for exponent
# np.log() syntax for log
```

```python
import numpy as np
arr=np.array([1,5,6,4])
print ("exponentiation (e^x)",np.exp(arr))
print("logarithm (ln)",np.log(arr))
```

```python
#matrix operations using np.dot(a,b) for multiplication
# np.transpose(a) swaps rows and columns
```

```python
matrix_a=np.array([[1,3],[4,5]])
matrix_b=np.array([[7,8],[5,6]])
print("matrix mutliplication :",np.dot(matrix_a,matrix_b))
print("original matrix:",matrix_a)
print("after transposition:",np.transpose(matrix_a))
```

```python
# calulating future value of investment
# FV=PMT x
```

```python
monthly_contribution=1000
annual_interest_rate=5.4
years=10

monthly_interest_rate=(annual_interest_rate/12)/100
n=years*12

future_value=monthly_contribution*((1+monthly_interest_rate)**n-1)/monthly_interest_rate
print("future values of investment:",round(future_value,2))
```

```python
temp=np.array([22,33,12,55,64])
rainfall=np.array([5,0,9,12,33])
hot_days=temp>30
rainy_days=rainfall>0
print("hot" ,temp>25)
print("rainy day",rainfall>10)
print("hot and rainy",temp & rainfall)
print("hot or rainy:",temp | rainfall)
```
